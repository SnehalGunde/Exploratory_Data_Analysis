## IRIS DATASET- EDA

```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns available datasets
sns.get_dataset_names()
```

```python
# extracting iris dataset
iris=sns.load_dataset("iris")
iris.head()
```

```python
#function to describe dataset
def describe_dataset(df):
    print("The shape of data set is : " ,df.shape)
    print('\n')
    print(df.describe())
    print('\n')
    print(df.info())
```

```python
#describe data set
describe_dataset(iris)
```

```python
iris.species.value_counts()
```

```python
#data cleaning

def data_cleaning(df):
    #check null values in dataset
    print(iris.isnull().sum()/100)
    
def uniquecol(df, colname):
    print(df[colname].nunique())
    print(df[colname].unique())
```

```python
data_cleaning(iris)
```

Dataset has no null values 


```python
#unique values in species column
uniquecol(iris,'species')
    
```

Sepal Width has outliers. Remove outliers

```python
q_low = iris["sepal_width"].quantile(0.01)
q_hi  = iris["sepal_width"].quantile(0.98)

df_filtered = iris[(iris["sepal_width"] < q_hi) & (iris["sepal_width"] > q_low)]
```

```python
df_filtered.sepal_width.describe()
```

```python
sns.boxplot(data=df_filtered,x='sepal_width')
plt.show()
```

```python
df_filtered.describe()
```

```python
df_filtered.info()
```

```python
#examine 'Species column'
print("Filtered data set"+"\n",df_filtered.species.value_counts())
print('\n')
print("Iris data set"+"\n",iris.species.value_counts())
```

```python
df_filtered.species.value_counts().plot.pie()
```

```python
#plotting heat map for irisdataset
plt.figure(figsize=(7,4)) 
sns.heatmap(iris.corr(),annot=True,fmt="f",cmap="YlGnBu")
plt.show()
```

```python
#plotting heat map for filtered dataset
plt.figure(figsize=(7,4)) 
sns.heatmap(df_filtered.corr(),annot=True,fmt="f",cmap="YlGnBu")
plt.show()
```
